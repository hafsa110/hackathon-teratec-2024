### Temporal profile

```python
import matplotlib.pyplot as plt
import os.path

from data_manip.formats.mascaret_file import MascaretFile
from postel.plot1d import plot1d


# Selection to display
courlis_file = MascaretFile(os.path.join(os.environ['HOMETEL'], 'examples', 'courlis', 'Garonne', 'mascaret0.listingcourlis'))
varnames = ['Concentration en vase', 'Concentration en sable']
reach_id = 1  # first reach
time_index = -1  # last frame
reach = courlis_file.reaches[reach_id]

fig, ax = plt.subplots(figsize=(10,8))

for varname in varnames:
    var_pos = courlis_file.get_position_var(varname)
    values = courlis_file.get_values_at_reach(time_index, reach_id, var_pos)
    plot1d(ax, reach.get_section_pk_list(), values,
           plot_label='%s' % varname,
           x_label='Distance (m)',
           y_label='Concentration (g/l)')

# Displaying legend
ax.legend()

# Showing figure
plt.show()

del plt
del courlis_file
```

### Longitudinal profile

```python
import matplotlib.pyplot as plt
import os.path

from data_manip.formats.mascaret_file import MascaretFile
from postel.plot1d import plot1d


# Selection to display
courlis_file = MascaretFile(os.path.join(os.environ['HOMETEL'], 'examples', 'courlis', 'Soni', 'resu.listingcourlis'))
varname = 'Cote du fond'
reach_id = 1  # first reach
reach = courlis_file.reaches[reach_id]
sections = reach[:5]  # First 5 sections of requested reach

# Build plot
fig, ax = plt.subplots(figsize=(10, 8))

var_pos = courlis_file.get_position_var(varname)
for section in sections:
    values = courlis_file.get_series(reach_id, section.id, [var_pos])
    plot1d(ax, courlis_file.times, values,
           plot_label='%s' % section.name,
           x_label='Temps (s)',
           y_label='Z fond (m)')

# Displaying legend
ax.legend()

# Showing figure
plt.show()

del plt
del courlis_file
```

### Extract a bottom/layer state from a ptravers file to a Courlis geometry file

It can be useful to restart a calculation but not from the last timestep or to make a Mascaret simulation on a particular state of bottoms with classical mode or with Mascaret Python API mode.

It can be done with `extract_ptravers_res_to_geoc.py` python sript (see [Courlis scripts to extract results and convert result file](../pretel/courlis.ipynb))

import os.path
from data_manip.formats.mascaret_file import MascaretFile

courlis_file = MascaretFile(os.path.join(os.environ['HOMETEL'], 'examples', 'courlis', 'Garonne', 'mascaret0.ptravers'))




```python
import os.path
from data_manip.formats.mascaret_file import MascaretFile, Reach
from data_manip.formats.mascaretgeo_file import MascaretGeoFile

courlis_ptravers = MascaretFile(os.path.join(os.environ['HOMETEL'], 'examples', 'courlis', 'Garonne', 'mascaret0.ptravers'))

# Third time frame of the result file is selected (time frame in MascaretFile begins by 0)
record = 2
output_name = os.path.join(os.environ['HOMETEL'],
                           'examples',
                           'courlis',
                           'Garonne',
                           'Garonne_time_'+str(courlis_ptravers.times[record])+'s_rep.geoC')
print(output_name)

# minus 4 because there are 4 variables in addition to layers elevation
# minus 1 because it is elevation of layer, river bottom and hard bottom, not thicknesses of layers
nlayers = courlis_ptravers.nsectionvar - 4 - 1

# Keeping only cross profile points abscissa and elevations (layers, river bed bottom and hard bottom)
section_vars_indexes = [i for i in range(0, nlayers + 2)]

# Extracting section variables
_, layers_elevation = courlis_ptravers.get_values(
                                                  record,
                                                  get_section_values=True,
                                                  section_vars_indexes=section_vars_indexes
                                                  )

# Building a MascaretGeoFile
courlis_geo = MascaretGeoFile(output_name, mode='write')

# Only one reach in Courlis
reach = Reach(1, "Reach")
# Enabling and creating layers in MascaretGeoFile (reach level)
courlis_geo.has_layers = True
courlis_geo.nlayers = nlayers
for i in range(nlayers+1):
    courlis_geo.layer_names.append(courlis_ptravers.section_varnames_dict['names'][i + 1])

# Creating section (extracted from ptravers file) with layers
for i in range(courlis_ptravers.nsections):
    section = courlis_ptravers.reaches[1].sections[i+1]
    dist = layers_elevation[1][i][0]
    z_list = layers_elevation[1][i][1]
    section.set_points_from_trans(dist, z_list)

    for j in range(nlayers):
        thickness_table = layers_elevation[1][i][j+1] - layers_elevation[1][i][j+2]
        section.add_layer_from_thickness(thickness_table)

    reach.add_section(section)

# Adding the reach constructing to the MascaretGeoFile
courlis_geo.add_reach(reach)

# Exporting the Courlis geometry file (.geoC)
courlis_geo.save(output_name)
```

```python
with open(output_name) as f:
    print(f.read())
```
